```python
import requests
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/news/news_read.naver?article_id=0000894224&office_id=366&mode=mainnews&type=&date=2023-04-17&page=1'
res=requests.get(url)
html=res.text
html
```

```python
soup=BeautifulSoup(html,'html.parser')
soup
post_list =[]
post_list=soup.find_all('div',{'id':"content"})
post_list
```

```python
post_list[0].getText()
```

```python
title_list=[]
subtitle_list = []
date_list =[]

for post in post_list:
  # print(post)
  # print('-----')
  title = post.find('h2',{'class':"post-title"}).text.strip()
  subtitle = post.find('h3').text.strip()
  date = post.find('p').text.strip()
  title_list.append(title)
  subtitle_list.append(subtitle)
  date_list.append(date)
```
