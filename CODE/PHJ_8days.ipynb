## Q24.
<div class="alert alert-success">
[문제24]     
    
smoker 여부에 따라, group_val 컬럼에 '흡연자', '비흡연자' 표시하기
</div>

```python
smgu = lambda sm : ('비흡연자') if(sm=='no') else ('흡연자')
df['group_val']=df['smoker'].apply(smgu)
df.head()
# weight2 = lambda man, height : (height - 100) * 0.9 if(man) else (height - 100)
```

```python
df.drop(['group_val'], axis=1, inplace=True)
```

```python
df2 = pd.DataFrame({'group_val' : ['흡연자', '비흡연자']}, index =['yes', 'no'])
```

```python
df2
```

```python
pd.merge(df, df2, left_on='smoker', right_index = True, how = 'outer' )
```

```python
df['group_val']=df['smoker'].replace(['yes','no'],['흡연자','비흡연자'])
df.head()
```
