```python
# ticker(주식번호) 와 종목명
df_tic = pd.DataFrame({'ticker':[],'tic_name':[]})

for ticker in stock.get_market_ticker_list(date='20230410'):
  df_tic.loc[len(df_tic)]=[ticker,stock.get_market_ticker_name(ticker)]

df_tic.tail()
```

```python
# ticker(주식번호) 와 인덱스(종목구분)
df_ind = pd.DataFrame({'tic_index':[],'tic_index_name':[],'ticker':[]})

for tic_ind in stock.get_index_ticker_list(date='20230410',market='KOSPI'):
  tic_ind_list=stock.get_index_portfolio_deposit_file(tic_ind)
  i=0
  while i < len(tic_ind_list):
    df_ind.loc[len(df_ind)]=[tic_ind, stock.get_index_ticker_name(tic_ind),tic_ind_list[i]]
    i=i+1

df_ind.tail()
```

```python
# ticker(주식번호) 와 인덱스(종목구분) 
df_ind_tic=pd.merge(df_ind, df_tic, on='ticker', how = 'left' )
df_ind_tic.tail()
```
